```
import numpy as np
import matplotlib.pyplot as plt
from sympy import *
```

```
cost_profit = 25
cost_non_profit = 250 
budget = 2500

profit_min = 0
profit_max = budget / cost_profit

non_profit_min = 0
non_profit_max = budget / cost_non_profit
```

```
def n_profit(n_non_profit, budget):
    return (budget - 250 * n_non_profit) / 25

def n_non_profit(profit, budget):
    return (budget - 25 * profit) / 250
```

```
profit_x = np.linspace(profit_min, profit_max, 100)
non_profit_y = n_non_profit(profit_x, budget)

plt.figure(figsize=(10,5))
plt.plot(profit_x, non_profit_y)
plt.xlabel('Number of Profit inspections')
plt.ylabel('Number of Non-Profit inspections')
plt.title('Possible ways of spending the budget')
plt.show()
```

```
def revenues(profit, non_profit):
    return profit**(3/4) * non_profit**(1/4) * 7
```

```
from mpl_toolkits.mplot3d import Axes3D
profit_axis = np.linspace(profit_min, profit_max, 100)
non_profit_axis = np.linspace(non_profit_min, non_profit_max, 100)
profit_grid, non_profit_grid = np.meshgrid(profit_axis, non_profit_axis)

fig = plt.figure()
ax = fig.gca(projection='3d')

ax.plot_surface(profit_grid, non_profit_grid, revenues(profit_grid,non_profit_grid))

ax.plot(non_profit_y, profit_x, linewidth = 5, color = 'r')

ax.set_xlabel('Number of profit visits')
ax.set_ylabel('Number of non profit visits')
ax.set_title('Possible ways of spending the budget')
plt.show()

```

```
fig, (ax_l, ax_r) = plt.subplots(1, 2, figsize = (15, 5))


profit_axis = np.linspace(profit_min, profit_max, 100)
non_profit_axis = np.linspace(non_profit_max, non_profit_min, 100)
profit_grid, non_profit_grid = np.meshgrid(profit_axis, non_profit_axis)
    
im = ax_l.imshow(revenues(profit_grid, non_profit_grid), aspect = 'auto', extent=[profit_min, profit_max, non_profit_min, non_profit_max])
ax_l.plot(profit_axis, n_non_profit(profit_axis, 2500), 'r')
ax_l.set_xlabel('Number of profit visits')
ax_l.set_ylabel('Number of non_profit visits')
ax_l.set_title('Possible ways of spending the budget')


# The contours are showing how the intersection looks like

profit_axis = np.linspace(profit_min, profit_max)
non_profit_axis = np.linspace(non_profit_min, non_profit_max)
profit_grid, non_profit_grid = np.meshgrid(profit_axis, non_profit_axis)

im2 = ax_r.contour(revenues(profit_grid,non_profit_grid), extent=[profit_min, profit_max, non_profit_min, non_profit_max])
ax_r.plot(profit_axis, n_non_profit(profit_axis, 2500), 'r')
ax_r.set_xlabel('Number of profit visits')
ax_r.set_ylabel('Number of non_profit visits')
ax_r.set_title('Possible ways of spending the budget')

plt.colorbar(im,ax=ax_l)
plt.colorbar(im2,ax=ax_r)

plt.show()
```
